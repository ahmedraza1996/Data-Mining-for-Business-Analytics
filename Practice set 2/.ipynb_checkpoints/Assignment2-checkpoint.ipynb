```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set(style="whitegrid")
pd.set_option("display.max_columns" , None)
warnings.filterwarnings('ignore')
```

# PROBLEM 3.1a,  Shipments of Household Appliances: Line Graphs.

```python
shipment_df = pd.read_csv('ApplianceShipments.csv')
```

```python
#convert quarters to date
shipment_df['date'] = pd.to_datetime(shipment_df['Quarter'].str[-4:] + shipment_df['Quarter'].str[:2])
```

```python
shipment_df.head()
```

```python
shipment_df.info()
```

__a. Create a well-formatted time plot of the data using Python__

```python
fig, ax = plt.subplots(figsize=(15, 6))
#d = data[data['obs_id']==2]
g =sns.lineplot(shipment_df['date'], shipment_df['Shipments'] , marker='o')
ax.set_title('Shipments over Time', fontsize = 20, loc='center', fontdict=dict(weight='bold'))
ax.set_xlabel('Year', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_ylabel('Shipments', fontsize = 16, fontdict=dict(weight='bold'))
g.set(ylim=(3000,5500))
plt.tick_params(axis='y', which='major', labelsize=16)
plt.tick_params(axis='x', which='major', labelsize=16)
```

# PROBLEM 3.3, Laptop Sales at a London Computer Chain: Bar Charts and Boxplots.

```python
laptop_df = pd.read_csv('LaptopSalesJanuary2008.csv')
```

```python
laptop_df.head()
```

```python
laptop_df.info()
```

```python
grouped_df=laptop_df[['Store Postcode', 'Retail Price']].groupby(['Store Postcode']).mean().reset_index().sort_values('Retail Price',ascending=False)

grouped_df=grouped_df.rename({'Store Postcode': 'Store_Postcode','Retail Price':'Retail_Price' } , axis=1)
grouped_df['Retail_Price']= round(grouped_df['Retail_Price'],1)
grouped_df.reset_index(drop=True, inplace=True)
```

__a. Create a bar chart, showing the average retail price by store.
Which store has the highest average? Which has the lowest?__

```python
fig, ax = plt.subplots(figsize=(15, 6))
g=sns.barplot(x="Store_Postcode", y="Retail_Price", data=grouped_df)
for index, row in grouped_df.sort_values('Retail_Price',ascending=False).iterrows():
    g.text(index,row.Retail_Price,row.Retail_Price, color='black', ha="center")

ax.set_title('Avg. Retail Price by Store', fontsize = 20, loc='center', fontdict=dict(weight='bold'))
ax.set_xlabel('Store', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_ylabel('Retail Priceq', fontsize = 16, fontdict=dict(weight='bold'))
g.set(ylim=(450,500))
plt.tick_params(axis='y', which='major', labelsize=16)
plt.tick_params(axis='x', which='major', labelsize=16)
ax.tick_params(axis='x', rotation=90)
plt.show()
```

Highest Average Retail Price Shop: __N17 6QA__

Lowest Average Retail Price Shop: __W4 3PH__


__b. To better compare retail prices across stores, create side-byside boxplots of retail price by store. Now compare the
prices in the two stores from (a). Does there seem to be a
difference between their price distributions?__


```python
fig, ax = plt.subplots(figsize=(15, 6))
g = sns.boxplot(x="Store Postcode", y="Retail Price", data=laptop_df)
ax.set_title('Avg. Retail Price by Store', fontsize = 20, loc='center', fontdict=dict(weight='bold'))
ax.set_xlabel('Store', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_ylabel('Retail Priceq', fontsize = 16, fontdict=dict(weight='bold'))
plt.tick_params(axis='y', which='major', labelsize=16)
plt.tick_params(axis='x', which='major', labelsize=16)
ax.tick_params(axis='x', rotation=90)
plt.show()
```

```python
# To have better visibilty among the two stores
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
g = sns.boxplot(x="Store Postcode", y="Retail Price", data=laptop_df[(laptop_df['Store Postcode']=='N17 6QA') | (laptop_df['Store Postcode']=='W4 3PH')])
ax.set_title('Retail Price distribution', fontsize = 20, loc='center', fontdict=dict(weight='bold'))
ax.set_xlabel('Store', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_ylabel('Retail Priceq', fontsize = 16, fontdict=dict(weight='bold'))
plt.tick_params(axis='y', which='major', labelsize=16)
plt.tick_params(axis='x', which='major', labelsize=16)
ax.tick_params(axis='x', rotation=90)
plt.show()
```
